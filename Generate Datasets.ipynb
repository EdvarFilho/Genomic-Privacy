```python
import numpy as np
import pandas as pd
import random
```

```python
dataset1 = np.zeros([100, 20001], dtype = object)
dataset2 = np.zeros([100, 50001], dtype = object)
dataset3 = np.zeros([1000, 20001], dtype = object)
dataset4 = np.zeros([1000, 50001], dtype = object)
```

```python
totalCasos1 = int(0.5*dataset1.shape[0])
totalCasos2 = int(0.5*dataset2.shape[0])
totalCasos3 = int(0.5*dataset3.shape[0])
totalCasos4 = int(0.5*dataset4.shape[0])
print(totalCasos1, totalCasos2, totalCasos3, totalCasos4)
```

```python
for i in range(0, dataset1.shape[0]):
    dataset1[i,0:dataset1.shape[1]-1] = np.random.randint(3, size=dataset1.shape[1]-1)
    if(np.count_nonzero(dataset1[:,dataset1.shape[1]-1] == 'Caso')<totalCasos1):
        dataset1[i,dataset1.shape[1]-1] = random.choice(['Caso', 'Controle'])
    else:
        dataset1[i,dataset1.shape[1]-1] = 'Controle'
dataset1
```

```python
for i in range(0, dataset2.shape[0]):
    dataset2[i,0:dataset2.shape[1]-1] = np.random.randint(3, size=dataset2.shape[1]-1)
    if(np.count_nonzero(dataset2[:,dataset2.shape[1]-1] == 'Caso')<totalCasos2):
        dataset2[i,dataset2.shape[1]-1] = random.choice(['Caso', 'Controle'])
    else:
        dataset2[i,dataset2.shape[1]-1] = 'Controle'
dataset2
```

```python
for i in range(0, dataset3.shape[0]):
    dataset3[i,0:dataset3.shape[1]-1] = np.random.randint(3, size=dataset3.shape[1]-1)
    if(np.count_nonzero(dataset3[:,dataset1.shape[1]-1] == 'Caso')<totalCasos3):
        dataset3[i,dataset3.shape[1]-1] = random.choice(['Caso', 'Controle'])
    else:
        dataset3[i,dataset3.shape[1]-1] = 'Controle'
dataset3
```

```python
for i in range(0, dataset4.shape[0]):
    dataset4[i,0:dataset4.shape[1]-1] = np.random.randint(3, size=dataset4.shape[1]-1)
    if(np.count_nonzero(dataset4[:,dataset4.shape[1]-1] == 'Caso')<totalCasos4):
        dataset4[i,dataset4.shape[1]-1] = random.choice(['Caso', 'Controle'])
    else:
        dataset4[i,dataset4.shape[1]-1] = 'Controle'
dataset4
```

```python
df1 = pd.DataFrame(dataset1)
df1.to_csv("dataset1.csv", index = False)
df1.head()
```
